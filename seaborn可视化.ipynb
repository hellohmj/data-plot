```python
#忽略警告
import warnings
warnings.filterwarnings('ignore')
#导入seaborn库
import seaborn as sns
import matplotlib.pyplot as plt
#加载seaborn自带“tips”数据集
tips = sns.load_dataset("tips")
tips.head()
```

```python
#seaborrn的五种预定义主题风格
#darkgrid, whitegrid, dark, white, ticks
styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

#创建画布
plt.subplots(figsize=(18, 12))
#循环绘制五种风格的图表
for i, style in enumerate(styles):
    sns.set_style(style)
    plt.subplot(2, 3, i+1)
    sns.boxplot(x='day', y='total_bill', data=tips)
    plt.title(f'style: {style}')
    plt.xlabel('Total Bill')
    plt.ylabel('Tip($)') 
plt.savefig('seaborn_style.png',bbox_inches='tight',dpi=1200)
plt.tight_layout()
plt.show()
```

分类调色板

```python
#定义seaborn调色板
#deep, muted, pastel, bright, dark, colorblind
categorical_palettes = ['deep', 'muted', 'bright', 'pastel', 'dark', 'colorblind']
plt.figure(figsize=(15,10))
#循环绘制六种调色板的图表
for i, palette in enumerate(categorical_palettes):
 
    sns.set_palette(palette)
    plt.subplot(2, 3, i+1)
    sns.barplot(x='day', y='total_bill',hue='sex' ,data=tips)
    plt.title(f'palette: {palette}')
plt.savefig('seaborn_palette.png',bbox_inches='tight',dpi=600)
plt.tight_layout()
plt.show()
```

连续调色板

```python
#定义seaborn调色板
sequential_palettes = ['Blues', 'BuGn', 'GnBu', 'Greens', 'Oranges', 'OrRd', 'PuBu', 'PuRd', 'Purples', 'RdPu', 'Reds', 'YlGn', 'YlGnBu', 'YlOrBr', 'YlOrRd', 'afmhot', 'autumn', 'bone', 'cool', 'copper', 'gist_heat', 'gray', 'hot', 'pink', 'spring', 'summer', 'winter', 'cubehelix']

plt.figure(figsize=(28,30))
#循环绘制28种调色板的图表
for i, palette in enumerate(sequential_palettes):
    sns.set_palette(palette)
    plt.subplot(7, 4, i+1)
    sns.violinplot(x='day', y='total_bill', data=tips,palette=sns.color_palette(palette))
    
    plt.title(f'palette: {palette}')
    
plt.tight_layout()
plt.savefig('seaborn_sequential_palette.png',bbox_inches='tight',dpi=600)
plt.show()
```

发散调色板

```python
#定义seaborn调色板
diverging_palettes = ['RdBu', 'RdYlBu', 'PRGn', 'PiYG', 'BrBG', 'RdGy', 'PuOr', 'RdYlGn']

plt.figure(figsize=(18,18))
#循环绘制8种调色板的图表
for i, palette in enumerate(diverging_palettes):
    sns.set_palette(palette)
    plt.subplot(4, 2, i+1)
    sns.kdeplot(x='total_bill',hue='sex', data=tips, fill=True,palette=sns.color_palette(palette))
    plt.title(f'palette: {palette}')
plt.tight_layout()
plt.savefig('seaborn_diverging_palette.png',bbox_inches='tight',dpi=600)
plt.show()
```
